# Imports

```python
import pandas as pd
import re
```

# Preporocessing

## Reading the DataFrame

```python
books = pd.read_csv("books_for_home.csv")
books.head()
```

## Author Name

```python
def create_name(name):
    return name.split(", ")[1] + " " + name.split(", ")[0]
```

```python
books["Author"] = books["Author"].apply(create_name)
books.head()
```

## Extraing the Information About Book Name

### Series

```python
books["Book Name"][10]
```

```python
series_reg = re.compile(r"""
    \((.*), ## series name,
    """,
    re.VERBOSE)
series_reg.findall(books["Book Name"][10])[0]
```

```python
def get_series(name):
    series_reg = re.compile(r"""
    \((.*), ## series name,
    """,
    re.VERBOSE)
    series_name = series_reg.findall(name)
    if len(series_name) > 0:
        return series_name[0]
    else:
        return "Standalone"
```

```python
books["Series"] = books["Book Name"].apply(get_series)
books.head()
```

### Entry In Series

```python
entry_reg = re.compile(r"""
    \#(\d) ## word,
    """,
    re.VERBOSE)
entry_reg.findall(books["Book Name"][10])[0]
```

```python
def get_entry_in_series(name):
    entry_reg = re.compile(r"""
    \#(\d) ## word,
    """,
    re.VERBOSE)
    entry_in_series = entry_reg.findall(name)
    if len(entry_in_series) > 0:
        return int(entry_in_series[0])
    else:
        return 1
```

```python
books["Volume"] = books["Book Name"].apply(get_entry_in_series)
books.head()
```

### Name

```python
name_reg = re.compile(r"""
    (.*)\s\( ## word,
    """,
    re.VERBOSE)
name_reg.findall(books["Book Name"][10])
```

```python
def get_name(name):
    name_reg = re.compile(r"""
    (.*)\s\( ## word,
    """,
    re.VERBOSE)
    try:
        return name_reg.findall(name)[0]
    except:
        return name
```

```python
books["Name"] = books["Book Name"].apply(get_name)
books.head()
```

## Making DateTime Objects

```python
def create_datetime(date):
    try:
        return pd.to_datetime(date)
    except:
        return None
```

```python
books["Publication Date"] = books["Publication Date"].apply(create_datetime)
books["Date Started"] = books["Date Started"].apply(create_datetime)
books["Date Completed"] = books["Date Completed"].apply(create_datetime)
```

## Saving the DataFrame

```python
books.to_csv("Books_Final.csv", index=False)
```
